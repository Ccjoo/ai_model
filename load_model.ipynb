```python
#cut file name to breed
import re
  
def remove(list):
    pattern = '[0-9]'
    list = [re.sub(pattern, '', i) for i in list]
    return list
```

```python
df = pd.DataFrame(columns = ['Original Filename', 'Original', 'Name 1', 'Pred 1', 'Name 2', 'Pred 2', 'Name 3', 'Pred 3'])
temp = pd.DataFrame(columns = ['Original Filename', 'Original', 'Name 1', 'Pred 1', 'Name 2', 'Pred 2', 'Name 3', 'Pred 3'])
```

```python
#load file
for name in target:
    mylist = [f for f in glob.glob("./dataset/cropped/test/{0}/*.jpg".format(name))]

    from random import sample
    selected  = sample(mylist,100)
    
    for img in selected:

        row = []
        image_pil = tf.keras.preprocessing.image.load_img(img, target_size=(200, 200, 3))
        image = np.array(image_pil)
        test_image = image / 255
        test_image = test_image.reshape((-1,) + test_image.shape)

        y_pred = model.predict(test_image)

        class_per = y_pred[0]
        icecream = dict(zip(target, class_per))

        row.append(img[41:-4])
        row.append(name)
        mylist = remove(mylist)

        count = 0
        for w in sorted(icecream, key=icecream.get, reverse=True):
            row.append(w)
            row.append(icecream[w])

            count += 1
            if count == 3: break

        new_df = pd.DataFrame([row], columns =  ['Original Filename', 'Original', 'Name 1', 'Pred 1', 'Name 2', 'Pred 2', 'Name 3', 'Pred 3'])
        df = pd.concat([df, new_df], sort=False)
            
```
